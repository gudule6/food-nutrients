# Une petite exploration de openfood.ch...

Doc: https://www.openfood.ch/en/documentation

```python
import requests, pandas as pd
```

```python
response = requests.get('https://www.openfood.ch/api/v1/products')
if not response.ok:
    print('Error when loading the products', response)
```

```python
data = response.json()
```

## Liste très mal faite de nutriments pour faire un test

```python
nutrients_list = []
for item in data['data']:
    if item['attributes']['nutrients']:
        if len(item['attributes']['nutrients']) > 18:
            nutrients_list = item['attributes']['nutrients']
nutrient_names = [i['name-translations']['fr'] for i in nutrients_list]
```

```python
new_item = dict.fromkeys(nutrient_names)
```

```python
item_list = []
for item in data['data']:    
#    comprendre pourquoi beaucoup d'items n'ont pas de nom... (catégorie: p.ex. tableau de nutriments)
#    print(item['attributes']['name'], item['attributes'])
    name = item['attributes']['name']
    if 'fr' in item['attributes']['name-translations'].keys():
        name = item['attributes']['name-translations']['fr']
    if not name:
        continue
    new_item = dict.fromkeys(nutrient_names)
    
    new_item['Nom'] = name
    add_nutrient_list = {}
    for nutrient in item['attributes']['nutrients']:
        add_nutrient_list.update({nutrient['name-translations']['fr']:nutrient['per-hundred']})
    
    new_item.update(add_nutrient_list)
    new_item['Ingrédients'] = item['attributes']['ingredients']
    item_list.append(new_item)
```

```python
df = pd.DataFrame(item_list)
```
